# Creating the graph of downtown Pittsburgh

We will create graph for the transportation network of downtown Pittsburgh. This will includes roads, parking garages, and select landmark destinations. The end goal is to have a graph that is easy to use with the queueing network class `Queueing_network`.

The important parts of the cities infrastructure is taken from [OpenStreetMaps.org](http://www.openstreetmaps.org). We assume you have already downloaded the region you are interested in and have saved it locally on your computer. We've saved the region of downtown pittsburgh to `./cities/downtown_pittsbugh.osm`. We can convert the `osm` file to a `networkx` graph object using the `osm2nx` file. The following code does just that, and takes note of any how many lanes there are on a road, and whether there is a traffic light at that intersection.

Associated with each node is a dictionary of tags called data.

```python
import networkx   as nx
import osm2nx

G = osm2nx.read_osm('./cities/pittsburgh_downtown.osm')
for k in G.node.keys():
    G.node[k]['destination']    = 0
    G.node[k]['garage']         = 0
    G.node[k]['light']          = 0
    G.node[k]['cfcc']           = 0
    G.node[k]['lanes']          = 1
    G.node[k]['cap']            = 0
    if 'highway' in G.node[k]['data'].tags :
        highway = G.node[k]['data'].tags['highway']
        if highway == 'traffic_signals' :
            G.node[k]['light'] = 1
        if highway == 'crossing' and 'crossing' in G.node[k]['data'].tags :
            if G.node[k]['data'].tags['crossing'] == 'traffic_signals' :
                G.node[k]['light'] = 2
```

The graph `G` now contains the transportation network of downtown Pittsburgh. This includes walking and biking pathways throughout the city. We don't have much interest in walking and biking pathways so we will remove them. Edges roadways (i.e. edges that are not walkways) have the `tiger:cfcc` tag, so we need to make sure that we mark them accordingly. We'll remove the nodes that are not roadways later when we export the graph `G` to `graph_tool`.

```python
for key1 in G.edge.keys() :
    for key2 in G.edge[key1].keys() :
        try :
            if 'tiger:cfcc' in G.edge[key1][key2]['data'].tags :
                G.node[key2]['cfcc'] = 1
            if 'lanes' in G.edge[key1][key2]['data'].tags :
                G.node[key1]['lanes'] = int(G.edge[key1][key2]['data'].tags['lanes'])
        except TypeError : 
            continue
```

If it's not a roadway we have no use for it so we'll delete it's data dictionary.

```python
for key1 in G.edge.keys() :
    for key2 in G.edge[key1].keys() :
        if key2 == 'cfcc' or key2 == 'lanes' :
            continue
        try :
            del G.edge[key1][key2]['data']
        except KeyError or TypeError :
            try :
                del G.edge[key1][key2]['id']
            except KeyError or TypeError :
                continue
```

Another piece of information given by the osm file is the latitude and longitude of all parts of the network. We save this information and then delete the data dictionary since we no longer have any use for it.

```python
for key, value in G.node.items() :
    G.node[key]['lat']  = value['data'].lat
    G.node[key]['lon']  = value['data'].lon
    del G.node[key]['data']
```

Lastly, we need to know the locations of landmark destinations and garages. This information is typically missing from the osm file. To get this information I wrote a script that takes retrieves it from [parkpgh.org](http://www.parkpgh.org). Once we've retrieved this information we need to add destination and garage nodes to the graph.

```python
from get_destinations   import get_dests
g_dict, d_dict = get_dests()
c   = 0

for k in range(len(g_dict)) :
    G.add_node(c, g_dict[k])
    c  += 1

for k in range(len(d_dict)) :
    G.add_node(c, d_dict[k])
    c  += 1
```

Our prefered graphing module is graph_tool, not networkx, so now we're going load the graph into graph_tool.

```python
import graph_tool.all       as gt

nx.write_graphml(G, 'downtown_pitt.xml')
g   = gt.load_graph('downtown_pitt.xml', fmt='xml')
```

Before we continue, it's worth looking at the graph as it is now.

```python
pos = g.new_vertex_property('vector<double>')
for v in g.vertices() :
    pos[v]  = [g.vp['lat'][v], g.vp['lon'][v]]

a = gt.graph_draw(g, pos=pos, inline=True)
```

This appears like a decent graph but there are several things wrong with it. For starters, there are a lot of long trailing nodes that go off into the distance. Since I have no need for these nodes they need to be removed. As I mentioned above there are walking paths included in this graph and I have no use for them either.

Before we start removing nodes we need to change the graph from undirected to directed since the `Queueing_network` class only works with directed graphs, and the graph given to us with `osm2nx` was undirected. This means one-way streets will not be represented as one-way streets. Still, we need to turn the graph into a directed graph. We add edges so that every street is a two-way street.

```python
g.set_directed(True)
g2  = g.copy()
for e in g2.edges() :
    target1 = int(e.target())
    source1 = int(e.source())
    e1      = g.add_edge(source=target1, target=source1)
```

Now we delete walk-ways and biking paths from the graph. Since every roadway node has its `cfcc` property set to `1`, this is easy.

```python
keep    = g.new_vertex_property('bool')
for v in g.vertices() :
    if g.vp['cfcc'][v] :
        keep[v] = True
    else :
        keep[v] = False

g.set_vertex_filter(keep, inverted=False)
g.purge_vertices()
```

Even after removing walkways and such we still have lots of nodes that we don't find useful.

```python
a = gt.graph_draw(g, pos=pos, inline=True)
```

We removing these nodes using the code below. It's tailor made for this graph in particular.

```python
def remove_problem_points() :
    latlong = [[40.442696, -79.9871978], [40.441443, -79.9849928], 
               [40.4469285, -80.0038217], [40.447944, -80.0003869], 
               [40.4326597, -80.0035281], [40.4404269, -79.9912726],
               [40.4405301, -79.9913136], [40.4366812, -80.0009142]]
    to_remove   = True
    while to_remove :
        g2  = g.copy()
        to_remove   = False
        for v in g2.vertices() :
            for latlon in latlong :
                if g.vp['lat'][v] == latlon[0] and g.vp['lon'][v] == latlon[1] :
                    to_remove   = True
                    w   = g.vertex(int(v))
                    g.remove_vertex(w)
                    break
            if to_remove :
                break


def remove_tails() :
    to_remove   = True
    while to_remove :
        g2  = g.copy()
        to_remove   = False
        for v in g2.vertices() :
            a   = not g2.vp['destination'][v] == 1
            b   = not g2.vp['garage'][v] == 1
            if v.out_degree() <= 1 and a and b:
                to_remove   = True
                w   = g.vertex(int(v))
                g.remove_vertex(w)
                break


remove_problem_points()
remove_tails()
```

```python
a = gt.graph_draw(g, pos=pos, inline=True)
```

We now have the graph broken into connected components. We want the nodes in the largest connected compnent as well as garage and destination nodes.

```python
import numpy as np
comp        = gt.label_components(g)
components  = g.new_vertex_property('bool')

s   = np.argmax([np.sum(comp[0].a == k) for k in range(np.max(comp[0].a))])
for v in g.vertices() :
    if comp[0][v] == s or g.vp['destination'][v] == 1 or g.vp['garage'][v] == 1 :
        components[v] = True

g.set_vertex_filter(components, inverted=False)
g.purge_vertices()
```

```python
# This is just so that these property maps are bool property
# maps, as opposed to integer property maps.
destination = g.new_vertex_property('bool')
garage      = g.new_vertex_property('bool')
light       = g.new_vertex_property('bool')

for v in g.vertices() :
    destination[v]  = True if g.vp['destination'][v] else False
    garage[v]       = True if g.vp['garage'][v] else False
    light[v]        = True if g.vp['light'][v]  else False
    
g.vp['light']       = light
g.vp['destination'] = destination
g.vp['garage']      = garage
```

Lastly, we add properties to make the graph look pretty.

```python
col         = g.new_vertex_property('vector<double>')
fill        = g.new_vertex_property('vector<double>')
vsize       = g.new_vertex_property('double')
vpen        = g.new_vertex_property('double')
cap         = g.new_vertex_property('int')


for v in g.vertices() :
    vpen[v]         = 0.6
    vsize[v]        = 6.5
    fill[v]         = [0.0, 0.5, 1.0, 1.0]
    col[v]          = [0.0, 0.5, 1.0, 1.0]
    if light[v] :
        fill[v]     = [1.0, 0.135, 0.0, 1.0]
        col[v]      = [1.0, 0.271, 0.0, 1.0]
    if garage[v] :
        col[v]      = [0.133, 0.545, 0.133, 1.0]
        fill[v]     = [0.133, 0.545, 0.133, 1.0]
    if destination[v] :
        col[v]      = [0.282, 0.239, 0.545, 1.0]
        fill[v]     = [0.282, 0.239, 0.545, 1.0]

g.vp['pos']         = pos
g.vp['col']         = col
g.vp['fill']        = fill
g.vp['vsize']       = vsize
g.vp['vpen']        = vpen

edge_color          = g.new_edge_property('vector<double>')
control             = g.new_edge_property('vector<double>')
edge_width          = g.new_edge_property('double')
arrow_width         = g.new_edge_property('double')


for e in g.edges() :
    edge_color[e]   = [0.339, 0.3063, 0.3170, 0.390]
    edge_width[e]   = 1.25
    arrow_width[e]  = 6
    control[e]      = [0, 0, 0, 0]


g.ep['edge_color']  = edge_color
g.ep['edge_width']  = edge_width
g.ep['arrow_width'] = arrow_width
g.ep['control']     = control
```

```python
a = gt.graph_draw(g, pos=g.vp['pos'], inline=True,
        vertex_fill_color=g.vp['fill'], 
        vertex_size=g.vp['vsize'], 
        vertex_pen_width=g.vp['vpen'],
        edge_color=g.ep['edge_color'],
        edge_control_points=g.ep['control'],
        edge_marker_size=g.ep['arrow_width'],
        edge_pen_width=g.ep['edge_width'])
```

Destination nodes are purple, garage nodes are green, intersections with traffic lights are red-orange, and all other nodes are blue.

## Finishing touches

There are only a few more finishing touches that we need to make before the graph is usable. Since we know the actual location of each node in our graph, can figure out the actual length of each road in the city. Also, destination and garage nodes are not connected to the rest of the network at the moment so we need to connect them. First the edge lengths.

###  Adding road distances

We use the following function to calculate the distance between two points in the network. 

```python
def calculate_distance( latlon1, latlon2 ) :
    lat1, lon1  = latlon1
    lat2, lon2  = latlon2
    R           = 6371          # radius of the earth in kilometers
    dlon        = lon2 - lon1
    dlat        = lat2 - lat1
    a           = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * (np.sin(dlon/2))**2
    c           = 2 * np.pi * R * np.arctan2( np.sqrt(a), np.sqrt(1-a) ) / 180
    return c
```

Now we can calculate the lengths of each edge. Note that the lengths are in kilometers. To change the edge lengths to miles express the radius of the earth in miles.

```python
edge_length = g.new_edge_property('double')

for e in g.edges() :
    latlon1         = g.vp['pos'][e.target()]
    latlon2         = g.vp['pos'][e.source()]
    edge_length[e]  = np.round( calculate_distance( latlon1, latlon2 ), 3)
    
g.ep['edge_length'] = edge_length
```

### Connecting destinations and garages

The following code connects each destination and garage node to the graph. For each node, it adds an edge to the two closest nodes in the graph. We need to add the lengths of the edges that were just added to our property map of edge lengths.

```python
dists   = np.ones( (g.num_vertices(), g.num_vertices()) ) * np.infty
for v in g.vertices() :
    if g.vp['garage'][v] or g.vp['destination'][v] :
        for v2 in g.vertices() :
            if not g.vp['garage'][v2] and not g.vp['destination'][v2] :
                latlon1 = g.vp['pos'][v]
                latlon2 = g.vp['pos'][v2]
                d       = np.round( calculate_distance( latlon1, latlon2 ), 3)
                dists[int(v), int(v2)] = d

mnd = np.min(dists)
g2  = g.copy()
for v in g2.vertices() :
    if g.vp['garage'][v] or g.vp['destination'][v] :
        vi      = int(v)
        d       = dists[vi, :]
        n       = np.argmin(d)
        e1      = g.add_edge(source=vi, target=n )
        e2      = g.add_edge(source=n,  target=vi)

        g.ep['edge_length'][e1] = d[n]
        g.ep['edge_length'][e2] = d[n]

        d[n]    = np.infty
        n       = np.argmin(d)
        e1      = g.add_edge(source=vi, target=n )
        e2      = g.add_edge(source=n,  target=vi)

        g.ep['edge_length'][e1] = d[n]
        g.ep['edge_length'][e2] = d[n]

        if g.vp['garage'][v] :
            e1  = g.add_edge(source=vi, target=vi)
            g.ep['edge_length'][e1] = mnd / 10
```

There are some destinations and garages in the northside of the city that we're not interested in. We now remove them.

```python
keep_removing = True
while keep_removing :
    keep_removing = False
    g2  = g.copy()
    for v in g2.vertices() :
        if g2.vp['garage'][v] or g2.vp['destination'][v] :
            for e in v.out_edges() :
                if g.ep['edge_length'][e] > 0.36 :
                    g.remove_vertex(int(v))
                    keep_removing = True
                    break
        if keep_removing :
            break
```

Add properties to these new edges.

```python
destination     = g.new_edge_property('bool')
garage          = g.new_edge_property('bool')

for e in g.edges() :
    if g.vp['garage'][e.target()] and g.vp['garage'][e.source()]:
        garage[e]       = True
        edge_length[e]  = mnd / 10
    if g.vp['destination'][e.target()]:
        destination[e]  = True
        latlon1         = g.vp['pos'][e.target()]
        latlon2         = g.vp['pos'][e.source()]
        edge_length[e]  = np.round( calculate_distance( latlon1, latlon2 ), 3)
        
g.ep['garage']          = garage
g.ep['destination']     = destination
```

### Rotating the graph

The last thing we need to do is rotate the graph. For some reason, graph look like it has been rotated clockwise by 90 degrees so I rotate it by -90 degrees.

```python
a,b = -15,32
rotation_m  = np.zeros( (2,2) )
rotation_m[0,0] =  np.cos( a*np.pi/b )
rotation_m[0,1] = -np.sin( a*np.pi/b )
rotation_m[1,0] =  np.sin( a*np.pi/b )
rotation_m[1,1] =  np.cos( a*np.pi/b )

for v in g.vertices() :
    g.vp['pos'][v] = np.dot( rotation_m, g.vp['pos'][v] )
```

The final product.

```python
for e in g.edges() :
    if g.ep['garage'][e] :
        g.ep['edge_color'][e]   = [0, 0, 0, 0]
    else :
        g.ep['control'][e]      = [0, 0, 0, 0]
        g.ep['edge_color'][e]   = [0.339, 0.3063, 0.3170, 0.390]
        g.ep['edge_width'][e]   = 1.25
        g.ep['arrow_width'][e]  = 6

    
a = gt.graph_draw(g, pos=g.vp['pos'], inline=True,
        vertex_fill_color=g.vp['fill'], 
        vertex_size=g.vp['vsize'],
        edge_control_points=g.ep['control'],
        edge_marker_size=g.ep['arrow_width'],
        edge_pen_width=g.ep['edge_width'])
```

As before, destination nodes are purple, garage nodes are green, intersections with traffic lights are red-orange, and all other nodes are blue.

Finally, we delete properties that we no longer find useful before saving.

```python
del g.vp['_graphml_vertex_id'], g.vp['vsize'], g.vp['col']
del g.vp['lat'], g.vp['lon'], g.vp['cfcc'], g.vp['vpen']
del g.vp['id'], g.vp['url']
del g.ep['arrow_width'], g.ep['edge_color']
del g.ep['edge_width'], g.ep['control']
del g.ep['_graphml_edge_id']

g.save('downtown_pitt.xml', fmt='xml')
```
